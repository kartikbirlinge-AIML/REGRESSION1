```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("height-weight.csv")
df
```

```python
plt.scatter(df['Weight'],df['Height'])
```

```python
# use correlation by that we can get the idea of x and y
df.corr()
```

```python
# creating the independant and dependant features
# and make sure that indepenfant features is always in a dataframe

X=df[['Weight']]
X
```

```python
# and dependant feature can be in series or in a 1 dimensional array
Y=df['Height']
Y
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.25,random_state=42)
```

```python
# standardization

from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
X_test
```

```python

# applying linear regression

from sklearn.linear_model import LinearRegression

regression=LinearRegression(n_jobs=-1)
regression.fit(X_train,Y_train)
```

```python
#ploting training data

plt.scatter(X_train,Y_train)
```

```python
plt.scatter(X_train,Y_train)
plt.plot(X_train,regression.predict(X_train))
```

```python
# prediction of test_data
Y_pred=regression.predict(X_test)
```

```python
regression.predict(scaler.transform([[72]]))
```

```python
# Performance metrics
from sklearn.metrics import mean_absolute_error,mean_squared_error

mse=mean_squared_error(Y_test,Y_pred)
mae=mean_absolute_error(Y_test,Y_pred)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```
